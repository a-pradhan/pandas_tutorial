```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
from numpy.random import randn

# Stats
from scipy import stats

# Plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# make plots appear in notebook
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.info()
```

```python
tips.describe()
```

```python
# regression plot of total_bill vs tips
sns.lmplot('total_bill','tip', tips)
```

```python
# we use dictionaries to edit individual parts
sns.lmplot('total_bill','tip',tips,
           scatter_kws={'marker':'o','color':'indianred'},
           line_kws={'linewidth':1,'color':'blue'})
```

```python
# can plot higher order trend lines (default is single order
sns.lmplot('total_bill','tip',tips,order=4,
           scatter_kws={'marker':'o','color':'indianred'},
           line_kws={'linewidth':1,'color':'blue'})
```

```python
# if we don't want a regression line
sns.lmplot('total_bill','tip',tips,fit_reg=False,
           scatter_kws={'marker':'o','color':'indianred'},
           line_kws={'linewidth':1,'color':'blue'})
```

```python
# lmplot also works for discrete variables like percentage of the tip
tips.head()
```

```python
# make new column
tips['tip_pct'] = 100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_pct',data=tips)
```

```python
# add a jitter to make te graph easier to read
sns.lmplot('size','tip_pct',data=tips,x_jitter=.1)
```

```python
# estimate tendency for each bin(size of the party)
# more variance for a party size of 1 than for 2,3,4
sns.lmplot('size', 'tip_pct',data=tips,x_estimator=np.mean)
```

```python
# using x for male, using * for female, using a hue to separate between the two sexes
sns.lmplot('total_bill','tip_pct',data=tips,hue='sex',markers=['x','o'])
```

```python
# Does day affect how much the tip is?
sns.lmplot('total_bill','tip_pct',data=tips,hue='day')
```

```python
# seaborn also support local regression model fitting
sns.lmplot('total_bill','tip_pct',data=tips,lowess=True,line_kws={'color':'black'})
```

```python
#lmplot is using a lower level plot called regplot
sns.regplot('total_bill','tip_pct',data=tips)
```

```python
# figure of two sub plots
# 1 by 2, sharey means their spaced out,
fig, (axis1, axis2) = plt.subplots(1,2,sharey=True)
sns.regplot('total_bill','tip_pct',data=tips,ax=axis1)
sns.violinplot(y=tips['tip_pct'],x=tips['size'],palette='Reds_r',ax=axis2)
```
