```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
from numpy.random import randn

# Stats
from scipy import stats

# Plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
# create boxplot using seaborn
sns.boxplot([data1,data2], vert=True)
```

```python
sns.boxplot([data1,data2],whis=np.inf, vert=False)
```

```python
# Normal Distribution
data1 = stats.norm(0,5).rvs(100)
# two gamma distributions concatenated together
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                        -1*stats.gamma(5).rvs(50)])
data2
```

```python
sns.boxplot(data=[data1,data2],whis=np.infty)
```

```python
# combine a kde plot with a box plot = violin plot
sns.violinplot(data=[data1,data2])
```

```python
# specify bandwith for density plot in violin plot
sns.violinplot(data=data2,bw=0.01)
```

```python
# can include invidual points
# all lines are invidual data points
sns.violinplot(data=[data1,data2],inner='stick')
```
